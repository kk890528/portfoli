# 嘗試畫出目前武漢肺炎確診人數的熱度圖，表示全球目前最嚴重的區域

[資料來源](https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data?fbclid=IwAR1C6Cf_k5bpeEF5v1I9G-OlXeMZd0B1JreUw4ayc5tuOLYVSKvyunFaam8)

```python
df=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/03-05-2020.csv')
```

```python
df.head()
```

```python
import folium
from folium.plugins import HeatMap
```

```python
confirmed_loc = []
for i in range(len(df)):
    for j in range(df.loc[i,'Confirmed']):
        confirmed_loc.append([df.loc[i,'Latitude'],df.loc[i,'Longitude']])
```

```python
maps= folium.Map(location=[20,20],zoom_start=2,tiles='Stamen Toner')
HeatMap(confirmed_loc, radius=17).add_to(maps)

maps.save('maps.html')
```
