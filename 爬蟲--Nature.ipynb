```python
import requests
from bs4 import BeautifulSoup
import time
import random
import string
import numpy as np
import pandas as pd
import datetime
```

```python
def natures(x):
    time.sleep(random.uniform(1, 5))
    nat=requests.get(x)
    nature=BeautifulSoup(nat.text, 'html.parser')
    #標題
    titles=nature.find('title').text.split('|')[0]
    #日期
    try:
        date=nature.find('meta',attrs={'name':'dc.date'})['content']
    except:
        date='請檢查'
    #作者
    author=''
    for i in nature.find_all('meta',attrs={'name':'dc.creator'}):
        au=i['content']
        author+=f',{au}'
    author=author[1:]
    #分類
    try:
        try:
            sention=nature.find('li',attrs={'data-test':'article-category'}).text
        except:
            sention=nature.find('meta',attrs={'name':'dc.type'})['content']
    except:
        sention=''
    #內容
    content=''
    try:
        for i in nature.find_all('div',attrs={'class':"c-article-main-column u-float-left"})[0].find_all('p'):
            content+=i.text
            words=len(content.split(' '))-120
    except:
        c=nature.find_all('div',attrs={'class':'c-article-section__content'})
        for i in range(0,len(c)):
            try:
                if c[i]['id']=='Bib1-content':
                    break
                else:
                    content+=c[i].text
            except:
                content+=c[i].text
        words=len(content.split(' '))
    
    #圖片
    n_p=len(nature.find_all('figure'))
    #影片

    v_p=len(nature.find_all('div',attrs={'class':'ratio--16-9'}))
    v_p
    if v_p !=0:
        all_p=f'{n_p}（+{v_p}影片）'
    else:
        all_p=n_p
    if 'podcast' in sention:
        content='podcast新聞，自行檢查'
        all_p='podcast新聞，自行檢查'
    if 'video' in 'sention':
        content='video新聞，自行檢查'
        all_p='video新聞，自行檢查'
    if words==1:
        content='內容請檢查'
    return [titles,content,words,sention,date,author,all_p]
```

```python
https=[]
for i in range(1,9):
    r=requests.get(f'https://www.nature.com/search?q=Vaccine&order=date_desc&date_range=last_year&page={i}')
    soup=BeautifulSoup(r.text, 'html.parser')
    for i in soup.find_all('h2',attrs={'class':'h3 extra-tight-line-height'}):
        h=i.find("a")["href"]
        http=f'https://www.nature.com{h}'
        https.append(http)
    time.sleep(random.uniform(1, 5))
```
